```python
from pandas import read_csv
import numpy
```

```python
dataset = read_csv('Transactions.csv')
dataset
```

```python
num_cols = len(dataset.columns)
num_rows = len(dataset.index)
test_data = dataset
print(num_cols)
print(num_rows)
```

```python
totalMissing = []
for x in range(num_cols):
    test = dataset.iloc[:,x]
    true_index = test.isna()
    totalMissing.append([dataset.columns[x], len(test[true_index.values])])
```

```python
for row in totalMissing:
    print('    '.join([str(elem) for elem in row]))
```
