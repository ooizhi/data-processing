```python
import pandas as pd
import numpy
from pandas import read_csv
```

```python
dataset = read_csv('Transactions.csv', header=None)
dataset
```

```python
num_cols = len(dataset.columns)
test_data = dataset
num_cols
```

```python
dataset.fillna('0',inplace=True)

for x in range(num_cols):
    test = pd.Series(dataset.iloc[:,x])
    test = test.astype(str)
    true_index1 = test.str.isalpha()
    count1 = len(test[true_index1])
    true_index2 = test.str.isnumeric()
    count2 = len(test[true_index2])
    if count1>count2:
        print("is String type")
        test[true_index2.values] = "NaN"
        test_data[x] = test
    elif count1<count2:
        print("is integer type")
        test[true_index1.values] = "NaN"
        test_data[x] = test
```
