## import numpy, pandas, matplotlib.pyplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import read_csv
```

## read the csv file data and store in df array

```python
df = read_csv('test1.csv', header=None)
df
```

## display graph

```python
plt.hist(df.values)
plt.show()
```

## Formula to check outlier

```python
lower_bound = 0.1 
upper_bound = 0.95
res = df.quantile([lower_bound, upper_bound])
res
```

## to show the no outlier 

```python
true_index = (res.loc[lower_bound] < df) & (df < res.loc[upper_bound])
true_index
```

## to show the outlier

```python
false_index = ~true_index
false_index
```

## to show the no outlier data

```python
df.values[true_index.values]
```

## to find the median

```python
mid = np.median(df.values[true_index.values])
mid
```

## change the outlier data to median value

```python
df[false_index.values] = mid
mid
```
