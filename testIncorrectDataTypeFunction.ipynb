```python
import pandas as pd
import numpy
from pandas import read_csv
import re
```

```python
dataset = read_csv('test4.csv')
dataset
```

```python
num_cols = len(dataset.columns)
test_data = dataset
num_cols
```

```python
dataset.fillna("",inplace=True)

for x in range(num_cols):
    test = pd.Series(dataset.iloc[:,x])
    test = test.astype(str)
    true_index1 = test.str.contains('[^0-9|.]')
    count1 = len(test[true_index1])
    true_index2 = test.str.isnumeric() | test.str.contains('^\d[.]\d$')
    count2 = len(test[true_index2])
    if count1>count2:
        print("is String type", count1, count2)
        test[~true_index1.values] = None
    elif count1<count2:
        print("is integer type", count1, count2)
        test[~true_index2.values] = None
    test_data.iloc[:,x] = test
```
